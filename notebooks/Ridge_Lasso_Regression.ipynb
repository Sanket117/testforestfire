# Algerian Forest Fires Dataset

## About Dataset

The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attributes and 1 output attribute (class)
The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classe

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header =1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## Clean the Dataset

```python
## Missing values
dataset[dataset.isnull().any(axis = 1)]
```

### The daset is converted into two sets, based on region from 122th index, we can make a new column based on the region

```python
dataset.loc[:122,"Region"] =0
dataset.loc[122:,"Region"] =1
df =dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Remove the null values
df = df.dropna().reset_index(drop =True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df =df.drop(122).reset_index(drop= True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
## Change the required columns as integer data types
```

```python
df[['month','day','year','Temperature','RH','Ws']] = df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

### changing the other columns to float data datatype

```python
object = [features for features in df.columns if df[features].dtypes =='O']
```

```python
for i in object:
    if i!='Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
## Let save the clean dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index =False)
```
